```python
posts =  test_df.TitleText.to_list()
```

```python
segment_and_analyze_sentiment_distilbert_majority_labels(posts[1])
```

```python
segment_and_analyze_sentiment_distilbert_majority_labels_2(posts[0])
```

```python
sentiments_new = list()
cnt = 0
for post in posts:
  print(cnt)
  sentiment = segment_and_analyze_sentiment_distilbert_majority_labels_2(post)
  cnt+=1
  sentiments_new.append(sentiment['label'])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.metrics import classification_report


# Create a confusion matrix
y_true = test_df["ground_truth"].tolist()
y_pred = sentiments_new

# Calculate accuracy, precision, recall, and f1-score
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='micro')
recall = recall_score(y_true, y_pred, average='micro')

print("---------------"+"sentiments_new"+"--------------")
# Print the results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)

# Print the confusion matrix
print(confusion_matrix(y_true, y_pred))
print(classification_report(y_true, y_pred))
```
